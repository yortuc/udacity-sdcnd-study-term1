```python
import tensorflow as tf
```

```python
# create constant
hello_constant = tf.constant("Hello world")

with tf.Session() as sess:
    # run the hello_constant operation in session
    output = sess.run(hello_constant)
    print(output)
```

```python
# placeholder
hello_plc = tf.placeholder(tf.string)

with tf.Session() as sess:
    output = sess.run(hello_plc, feed_dict={hello_plc: "hello world!"})
    print(output)
```

```python
# varialbe initialization

n_labels = 5
n_features = 120

w = tf.Variable( tf.truncated_normal((n_features, n_labels)) )
b = tf.Variable( tf.zeros(n_labels) )

init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    w_val = sess.run(w)
    b_val = sess.run(b)
    print(w_val, b_val)
```
